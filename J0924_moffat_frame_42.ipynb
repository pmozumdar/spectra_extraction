```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from astropy.table import Table
from astropy.modeling import models
from keckcode_test.keckcode.esiredux import esi2d, esi1d
from specim_test.specim.specfuncs import spec2d, spec1d, echelle1d
```

```python
file_loc = '../galaxy_spectra/ESI_data/*'
```

```python
spec2d_file_list = []
var2d_file_list = []
#out_file = []
for i,p in enumerate(glob.glob(file_loc)):
    if 'J0924' in p:
        if 'bgsub.fits' in p:
            spec2d_file_list.append(p)
        elif 'var.fits' in p:
            var2d_file_list.append(p)
```

```python
spec = esi2d.Esi2d(spec2d_file_list[0], varfile=var2d_file_list[0])
print(spec2d_file_list[0])
print(var2d_file_list[0])
```

```python
pixrange = [(1900, 2400)]
for sp, info in zip(spec, spec.ordinfo):
    title = 'Spatial profile for ' + str(info['name'])
    sp.spatial_profile(pixrange=pixrange[0], title=title)
    plt.show()
    sp.initial_model()
```

```python
def tie_pos2_moffat(model):
    #print(pixscale)
    return model.x_0_2 + (1.81 / pixscale)  #(1.87 / pixscale)  

def tie_pos1_moffat(model):
    #print(pixscale)
    return model.x_0_2 + (0.87 / pixscale) #(0.886 / pixscale)

def tie_gamma_moffat(model):
    return model.gamma_2

def tie_alpha_moffat(model):
    return model.alpha_2
    
for sp, info in zip(spec, spec.ordinfo):
    if sp.mod0 is not None:
        pixscale = info['pixscale']
        sp.mod0[1].mean.tied = tie_pos1_moffat
        #sp.mod0[3].mean.tied = tie_pos2_gauss
        #sp.mod0[1].stddev.tied = tie_stddev_gauss
        sp.mod0[3].x_0.tied = tie_pos2_moffat
        #sp.mod0[2].x_0.tied = tie_pos1_moffat
        sp.mod0[3].gamma.tied = tie_gamma_moffat
        sp.mod0[3].alpha.tied = tie_alpha_moffat
        sp.refined_model(sp.mod0)
```

```python
outspec = spec.extract_all(method='cdf', plot_profiles=False, plot_traces=True, plot_extracted=False,
                           stepsize=25, pixrange=[0, -1, 2200, -1], level=-20, fit_dict={10 : [0, 2000]}, 
                           extrange_dict={10 : [0, 2000]}, 
                           fitorder_dict={'common':{'mean_1': 4, 'x_0_2':4, 'x_0_3':4}})
```

```python
outspec[1].save_multi('esi1d_J0924_42_moffat_primary')
outspec[0].save_multi('esi1d_J0924_42_moffat_secondary')
outspec[2].save_multi('esi1d_J0924_42_moffat_tertiary')
```
