```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from keckcode_test.keckcode.esiredux import esi2d, esi1d
from specim_test.specim.specfuncs import spec2d, spec1d, echelle1d
```

```python
file_loc = '../galaxy_spectra/ESI_data/*'
```

```python
spec2d_file_list = []
var2d_file_list = []
#out_file = []
for i,p in enumerate(glob.glob(file_loc)):
    if 'J0924' in p:
        if 'bgsub.fits' in p:
            spec2d_file_list.append(p)
        elif 'var.fits' in p:
            var2d_file_list.append(p)
```

```python
spec = esi2d.Esi2d(spec2d_file_list[0], varfile=var2d_file_list[0])
print(spec2d_file_list[0])
print(var2d_file_list[0])
```

```python
pixrange = [(1900, 2400)]
for sp, info in zip(spec, spec.ordinfo):
    title = 'Spatial profile for ' + str(info['name'])
    sp.spatial_profile(pixrange=pixrange[0], title=title)
    plt.show()
    sp.initial_model()
```

```python
def tie_pos1_moffat_1(model):
    return model.x_0_2 + (0.87 / spec.ordinfo[0]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_1(model):
    return model.x_0_2 + (1.81 / spec.ordinfo[0]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_2(model):
    return model.x_0_2 + (0.87 / spec.ordinfo[1]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_2(model):
    return model.x_0_2 + (1.81 / spec.ordinfo[1]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_3(model):
    return model.x_0_2 + (0.87 / spec.ordinfo[2]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_3(model):
    return model.x_0_2 + (1.81 / spec.ordinfo[2]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_4(model):
    return model.x_0_2 + (0.87 / spec.ordinfo[3]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_4(model):
    return model.x_0_2 + (1.81 / spec.ordinfo[3]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_5(model):
    return model.x_0_2 + (0.87 / spec.ordinfo[4]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_5(model):
    return model.x_0_2 + (1.81 / spec.ordinfo[4]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_6(model):
    return model.x_0_2 + (0.87 / spec.ordinfo[5]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_6(model):
    return model.x_0_2 + (1.81 / spec.ordinfo[5]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_7(model):
    return model.x_0_2 + (0.87 / spec.ordinfo[6]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_7(model):
    return model.x_0_2 + (1.81 / spec.ordinfo[6]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_8(model):
    return model.x_0_2 + (0.87 / spec.ordinfo[7]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_8(model):
    return model.x_0_2 + (1.81 / spec.ordinfo[7]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_9(model):
    return model.x_0_2 + (0.87 / spec.ordinfo[8]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_9(model):
    return model.x_0_2 + (1.81 / spec.ordinfo[8]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_10(model):
    return model.x_0_2 + (0.87 / spec.ordinfo[9]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_10(model):
    return model.x_0_2 + (1.81 / spec.ordinfo[9]['pixscale'])  #(1.87 / pixscale) 

def tie_gamma_moffat(model):
    return model.gamma_2

def tie_alpha_moffat(model):
    return model.alpha_2

func_dict = {'1_1' : tie_pos1_moffat_1, '1_2' : tie_pos2_moffat_1,
             '2_1' : tie_pos1_moffat_2, '2_2' : tie_pos2_moffat_2,
             '3_1' : tie_pos1_moffat_3, '3_2' : tie_pos2_moffat_3,
             '4_1' : tie_pos1_moffat_4, '4_2' : tie_pos2_moffat_4,
             '5_1' : tie_pos1_moffat_5, '5_2' : tie_pos2_moffat_5,
             '6_1' : tie_pos1_moffat_6, '6_2' : tie_pos2_moffat_6,
             '7_1' : tie_pos1_moffat_7, '7_2' : tie_pos2_moffat_7,
             '8_1' : tie_pos1_moffat_8, '8_2' : tie_pos2_moffat_8,
             '9_1' : tie_pos1_moffat_9, '9_2' : tie_pos2_moffat_9,
             '10_1' : tie_pos1_moffat_10, '10_2' : tie_pos2_moffat_10
            }

for sp, info in zip(spec, spec.ordinfo):
    if sp.mod0 is not None:
        print(info['pixscale'])
        sp.mod0[1].mean.tied = func_dict['%d_%d' %(info['order'], 1)] #tie_pos1_moffat
        sp.mod0[3].x_0.tied = func_dict['%d_%d' %(info['order'], 2)] #tie_pos2_moffat
        sp.mod0[3].gamma.tied = tie_gamma_moffat
        sp.mod0[3].alpha.tied = tie_alpha_moffat
        sp.refined_model(sp.mod0)
```

```python
outspec = spec.extract_all(method='cdf', plot_profiles=False, plot_traces=True, plot_extracted=False,
                           stepsize=25, pixrange=[0, -1, 2200, -1], level=-20, fit_dict={10 : [0, 2000]}, 
                           extrange_dict={10 : [0, 2000]}, 
                           fitorder_dict={'common':{'mean_1': 4, 'x_0_2':4, 'x_0_3':4}})
```

```python
for sp, info in zip(spec, spec.ordinfo):
    if sp.mod0 is not None:
        a = 1.81/(sp.coarsepars['x_0_3'] - sp.coarsepars['x_0_2'])
        b = 0.87/(sp.coarsepars['mean_1'] - sp.coarsepars['x_0_2'])
        print('order: %d' %info['order'])
        print(np.max(a), np.min(a))
        print(np.max(b), np.min(b))
```

```python
for sp, info in zip(spec, spec.ordinfo):
    if sp.mod0 is not None:
        a = 1.81/(sp.fitpars['x_0_3'] - sp.fitpars['x_0_2'])
        b = 0.87/(sp.fitpars['mean_1'] - sp.fitpars['x_0_2'])
        print('order: %d' %info['order'])
        print(np.max(a), np.min(a))
        print(np.max(b), np.min(b))
```
