```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from keckcode_test.keckcode.esiredux import esi2d, esi1d
from specim_test.specim.specfuncs import spec2d, echelle1d, spec1d, echelle2d
```

```python
file_loc = 'B0631_ESI_spectra/*'
```

```python
spec2d_file_list = []
var2d_file_list = []
#out_file = []
for i, p in enumerate(glob.glob(file_loc)):
    file_name = p.split('/')[-1]
    if 'sp_order' in file_name:
        spec2d_file_list.append(p)
    elif 'var' in file_name:
        var2d_file_list.append(p)
```

```python
print(spec2d_file_list, var2d_file_list)
```

```python
spec_ord_6 = spec2d.Spec2d(inspec=spec2d_file_list[3], invar=var2d_file_list[3], logwav=True)
spec_ord_7 = spec2d.Spec2d(inspec=spec2d_file_list[4], invar=var2d_file_list[0], logwav=True)
spec_ord_8 = spec2d.Spec2d(inspec=spec2d_file_list[1], invar=var2d_file_list[2], logwav=True)
spec_ord_9 = spec2d.Spec2d(inspec=spec2d_file_list[0], invar=var2d_file_list[1], logwav=True)
spec_ord_10 = spec2d.Spec2d(inspec=spec2d_file_list[2], invar=var2d_file_list[4], logwav=True)
```

```python
spec_list= [spec_ord_6, spec_ord_7, spec_ord_8, spec_ord_9, spec_ord_10]

pixrange_list = [[2000, 2400], [3000, 3400], [1700, 2100], [1600, 2000], [650, 950]] 

for i, sp in enumerate(spec_list):
    sp.spatial_profile(pixrange=pixrange_list[i])
    plt.show()
    sp.initial_model()
```

```python
fitrange_list = np.array([[900, 4100], [0, 4100], [0, 4100], [0, 4100], [0, 4100]])

for i, sp in enumerate(spec_list):
    sp.find_and_trace(sp.mod0, stepsize=25, fitorder={'x_0_1':3, 'mean_1': 3, 'x_0_2':4, 'x_0_3':4},
                    fitrange=fitrange_list[i], polyorder=3)
    plt.show()
```

```python
extrange_list = np.array([[900, -1], [600, -1], [200, -1], [0, -1], [0, 2000]])

for sp, ext in zip(spec_list, extrange_list):
    sp.extract(method='modelfit', extrange=ext)
    plt.show()
```
