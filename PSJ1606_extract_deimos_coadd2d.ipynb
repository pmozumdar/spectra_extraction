```python
#spec_list = [spec1, spec2]
shifted_sp = []
shifted_var = []
shift_pixel = np.array([0.0, -1.57, -1.62])

new_sp = []
new_var = []
for i, sp in enumerate(spec_list):
    #print(sp[order-1].mod0, shift[i])
    nanmask = (sp.vardata <= 0) | (np.isnan(sp.vardata)) 
    nnan = nanmask.sum()
    if nnan > 0:
        print('Found %d NaNs in the two-dimensional variance '\
              'spectrum' % nnan)

        """ Replace the NaNs with a big value """
        sp.vardata[nanmask] = 1.e-5
    shifted_sp.append(shift(sp.data, [shift_pixel[i], 0], order=1))
    shifted_var.append(shift(sp.vardata, [shift_pixel[i], 0], order=1))
```

```python
plt.figure(figsize=(8, 6))
color = ['b', 'r', 'k']
label = ['frame_88', 'frame_90', 'frame_91']
for i, p in enumerate(shifted_sp):
    tmpdata = p[2400:2800, :]
    pflux = np.median(tmpdata, axis=0)
    x = np.arange(p.shape[1])
    plt.plot(x, pflux, color=color[i], drawstyle='steps', label=label[i])
plt.legend()
#plt.title('order %d' %od)
```

```python
wtsp = np.zeros(shifted_sp[0].shape)
wtsum = np.zeros(shifted_var[0].shape)

for s, v in zip(shifted_sp, shifted_var):
    wt = np.zeros(wtsp.shape)
    mask = np.isnan(v) | (v<=0)
    wt[~mask] = v[~mask]
    #print(sum(sum(np.isnan(wt) | (wt<=0))))
    wtsp += wt * s
    #print(sum(sum(wtsp==0)))
    wtsum += wt
    #print(sum(sum(wtsum==0)))

wtsp[wtsum == 0] = 0
wtsum[wtsum == 0] = 1
outsp = wtsp / wtsum
outvar = 1.0 / wtsum
```

```python
def tie_pos1_moffat(model):
     return model.x_0_5 - 14.4
            
def tie_pos2_moffat(model):
    return model.x_0_1 + 5.1 

def tie_pos3_moffat(model):
    return model.x_0_1 +  6.35

def tie_pos4_moffat(model):
    return model.x_0_1 +  7.7

def tie_gamma_moffat(model):
    return model.gamma_5

def tie_alpha_moffat(model):
    return model.alpha_5 


mod[1].x_0.tied = tie_pos1_moffat
mod[2].x_0.tied = tie_pos2_moffat
mod[3].x_0.tied = tie_pos3_moffat
mod[4].x_0.tied = tie_pos4_moffat
mod[2].gamma.tied = tie_gamma_moffat
mod[2].alpha.tied = tie_alpha_moffat
mod[3].gamma.tied = tie_gamma_moffat
mod[3].alpha.tied = tie_alpha_moffat
mod[1].gamma.tied = tie_gamma_moffat
mod[1].alpha.tied = tie_alpha_moffat
mod[4].gamma.tied = tie_gamma_moffat
mod[4].alpha.tied = tie_alpha_moffat
spec.refined_model(mod)
```
