```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from astropy.table import Table
from astropy.modeling import models
from keckcode_test.keckcode.esiredux import esi2d, esi1d
from specim_test.specim.specfuncs import spec2d, spec1d
```

```python
file_loc = '../galaxy_spectra/ESI_data/*'
```

```python
spec2d_file_list = []
var2d_file_list = []
#out_file = []
for i,p in enumerate(glob.glob(file_loc)):
    if 'J0924' in p:
        if 'bgsub.fits' in p:
            spec2d_file_list.append(p)
        elif 'var.fits' in p:
            var2d_file_list.append(p)
```

```python
spec = esi2d.Esi2d(spec2d_file_list[1], varfile=var2d_file_list[1])
print(spec2d_file_list[1])
print(var2d_file_list[1])
```

```python
spec.plot_2d()
```

```python
pixrange = [(1900, 2400)]
for sp, info in zip(spec, spec.ordinfo):
    title = 'Spatial profile for ' + str(info['name'])
    sp.spatial_profile(pixrange=pixrange[0], title=title)
    plt.show()
    sp.initial_model()
```

```python
def tie_pos1_gauss(model):
    #print(pixscale)
    return model.mean_1 + (0.8 / pixscale) #(0.886 / pixscale)

def tie_pos2_gauss(model):
    #print(pixscale)
    return model.mean_1 + (1.87 / pixscale)

def tie_stddev_gauss(model):
    return model.stddev_1
   
for sp, info in zip(spec, spec.ordinfo):
    if sp.mod0 is not None:
        pixscale = info['pixscale']
        #sp.mod0[2].mean.tied = tie_pos1_gauss
        sp.mod0[2].mean.min = sp.mod0[1].mean + (0.850 / pixscale)
        sp.mod0[2].mean.max = sp.mod0[1].mean + (0.890 / pixscale)
        sp.mod0[3].mean.min = sp.mod0[1].mean + (1.65 / pixscale)
        sp.mod0[3].mean.max = sp.mod0[1].mean + (1.95 / pixscale)
        #sp.mod0[3].x_0.tied = tie_pos2_moffat
        #sp.mod0[3].mean.tied = tie_pos2_gauss
        sp.mod0[3].stddev.tied = tie_stddev_gauss
        #sp.mod0[2].gamma.tied = tie_gamma_moffat
        #sp.mod0[2].alpha.tied = tie_alpha_moffat
        sp.refined_model(sp.mod0)
```

```python
outspec = spec.extract_all(method='cdf', plot_profiles=False, plot_traces=True, plot_extracted=False,
                          stepsize=25)
```
