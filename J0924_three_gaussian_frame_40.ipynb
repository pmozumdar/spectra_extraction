```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from keckcode_test.keckcode.esiredux import esi2d, esi1d
from specim_test.specim.specfuncs import spec2d, spec1d
```

```python
file_loc = '../galaxy_spectra/ESI_data/*'
```

```python
spec2d_file_list = []
var2d_file_list = []
#out_file = []
for i,p in enumerate(glob.glob(file_loc)):
    if 'J0924' in p:
        if 'bgsub.fits' in p:
            spec2d_file_list.append(p)
        elif 'var.fits' in p:
            var2d_file_list.append(p)
```

```python
spec = esi2d.Esi2d(spec2d_file_list[2], varfile=var2d_file_list[2])
print(spec2d_file_list[2])
print(var2d_file_list[2])
```

```python
pixrange = [(1900, 2400)]
for sp, info in zip(spec, spec.ordinfo):
    title = 'Spatial profile for ' + str(info['name'])
    sp.spatial_profile(pixrange=pixrange[0], title=title)
    plt.show()
    sp.initial_model()
```

```python
def tie_pos1_gauss_1(model):
    return model.mean_1 + (0.87 / spec.ordinfo[0]['pixscale']) #(0.886 / pixscale) 0.87

def tie_pos2_gauss_1(model):
    return model.mean_1 + (1.81 / spec.ordinfo[0]['pixscale']) #(1.87 / pixscale) 1.81

def tie_pos1_gauss_2(model):
    return model.mean_1 + (0.87 / spec.ordinfo[1]['pixscale']) #(0.886 / pixscale) 0.87

def tie_pos2_gauss_2(model):
    return model.mean_1 + (1.81 / spec.ordinfo[1]['pixscale']) #(1.87 / pixscale) 1.81

def tie_pos1_gauss_3(model):
    return model.mean_1 + (0.87 / spec.ordinfo[2]['pixscale']) #(0.886 / pixscale) 0.87

def tie_pos2_gauss_3(model):
    return model.mean_1 + (1.81 / spec.ordinfo[2]['pixscale']) #(1.87 / pixscale) 1.81

def tie_pos1_gauss_4(model):
    return model.mean_1 + (0.87 / spec.ordinfo[3]['pixscale']) #(0.886 / pixscale) 0.87

def tie_pos2_gauss_4(model):
    return model.mean_1 + (1.81 / spec.ordinfo[3]['pixscale']) #(1.87 / pixscale) 1.81

def tie_pos1_gauss_5(model):
    return model.mean_1 + (0.87 / spec.ordinfo[4]['pixscale']) #(0.886 / pixscale) 0.87

def tie_pos2_gauss_5(model):
    return model.mean_1 + (1.81 / spec.ordinfo[4]['pixscale']) #(1.87 / pixscale) 1.81

def tie_pos1_gauss_6(model):
    return model.mean_1 + (0.87 / spec.ordinfo[5]['pixscale']) #(0.886 / pixscale) 0.87

def tie_pos2_gauss_6(model):
    return model.mean_1 + (1.81 / spec.ordinfo[5]['pixscale']) #(1.87 / pixscale) 1.81

def tie_pos1_gauss_7(model):
    return model.mean_1 + (0.87 / spec.ordinfo[6]['pixscale']) #(0.886 / pixscale) 0.87

def tie_pos2_gauss_7(model):
    return model.mean_1 + (1.81 / spec.ordinfo[6]['pixscale']) #(1.87 / pixscale) 1.81

def tie_pos1_gauss_8(model):
    return model.mean_1 + (0.87 / spec.ordinfo[7]['pixscale']) #(0.886 / pixscale) 0.87

def tie_pos2_gauss_8(model):
    return model.mean_1 + (1.81 / spec.ordinfo[7]['pixscale']) #(1.87 / pixscale) 1.81

def tie_pos1_gauss_9(model):
    return model.mean_1 + (0.87 / spec.ordinfo[8]['pixscale']) #(0.886 / pixscale) 0.87

def tie_pos2_gauss_9(model):
    return model.mean_1 + (1.81 / spec.ordinfo[8]['pixscale']) #(1.87 / pixscale) 1.81

def tie_pos1_gauss_10(model):
    return model.mean_1 + (0.87 / spec.ordinfo[9]['pixscale']) #(0.886 / pixscale) 0.87

def tie_pos2_gauss_10(model):
    return model.mean_1 + (1.81 / spec.ordinfo[9]['pixscale']) #(1.87 / pixscale) 1.81

def tie_stddev_gauss(model):
    return model.stddev_1

func_dict = {'1_1' : tie_pos1_gauss_1, '1_2' : tie_pos2_gauss_1,
             '2_1' : tie_pos1_gauss_2, '2_2' : tie_pos2_gauss_2,
             '3_1' : tie_pos1_gauss_3, '3_2' : tie_pos2_gauss_3,
             '4_1' : tie_pos1_gauss_4, '4_2' : tie_pos2_gauss_4,
             '5_1' : tie_pos1_gauss_5, '5_2' : tie_pos2_gauss_5,
             '6_1' : tie_pos1_gauss_6, '6_2' : tie_pos2_gauss_6,
             '7_1' : tie_pos1_gauss_7, '7_2' : tie_pos2_gauss_7,
             '8_1' : tie_pos1_gauss_8, '8_2' : tie_pos2_gauss_8,
             '9_1' : tie_pos1_gauss_9, '9_2' : tie_pos2_gauss_9,
             '10_1' : tie_pos1_gauss_10, '10_2' : tie_pos2_gauss_10
            }
    
for sp, info in zip(spec, spec.ordinfo):
    if sp.mod0 is not None:
        print(info['pixscale'])
        sp.mod0[2].mean.tied = func_dict['%d_%d' %(info['order'], 1)] #tie_pos1_gauss_
        sp.mod0[3].mean.tied = func_dict['%d_%d' %(info['order'], 2)] #tie_pos2_gauss
        sp.mod0[3].stddev.tied = tie_stddev_gauss
        sp.refined_model(sp.mod0)
```
