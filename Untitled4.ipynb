```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from astropy.io import fits
from specim_test.specim.specfuncs import spec2d, spec1d, echelle1d

from astropy.modeling.models import Moffat1D
from astropy.convolution import convolve, Box1DKernel
```

```python
mf = Moffat1D(amplitude=1500, x_0=10, gamma=5, alpha=8)
rect_ap = Box1DKernel(7)
```

```python
x = np.linspace(2, 20, 100)
data = mf(x)
```

```python
convolved_mf = convolve(data, rect_ap)
```

```python
plt.plot(x, data, label='moffat')
plt.plot(x, convolved_mf, label='convoled_moffat')
plt.legend()
```
