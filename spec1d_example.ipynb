```python
from specim.specfuncs import spec1d
from astropy.io import fits
import numpy as np
```

```python
sp = fits.open('../RXJ1131_arc.fits')
```

```python
wave = np.arange(5215) * 0.5 + 3329.0
```

```python
sp_arc = spec1d.Spec1d(wav=wave[800:-800], flux=sp[0].data[800:-800])
```

```python
%matplotlib notebook
sp_arc.smooth(1) # The integer input is the number of box to be used in smoothing
sp_arc.mark_lines('abs', z=0.655, usesmooth=True)
```
