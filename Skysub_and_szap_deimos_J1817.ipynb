```python
import numpy as np
import matplotlib.pyplot as plt
import glob
from specim_test.specim.specfuncs import spec2d
from astropy.io import fits as pf
```

```python
file_loc = '../galaxy_spectra/DEIMOS_data/J1817_data/*'
```

```python
file_list = []
out_file_det2 = []
out_file_det6 = []

for i,p in enumerate(glob.glob(file_loc)):
    if 'spec2d' in p and 'J1817' in p:
        file_list.append(p)
        frame = p.split('_')[-3][:4]
        out_file_det2.append('J1817_DEIMOS_spectra/J1817_deimos_det2_'+ frame + '_clean.fits')
        out_file_det6.append('J1817_DEIMOS_spectra/J1817_deimos_det6_'+ frame + '_clean.fits')
```

## median image after sky subtraction

```python
## exposure of each frame
for fl in file_list:
    sp = pf.open(fl)
    #m = sp.slitinfo['objid']=='lens'
    frame = fl.split('_')[-3][:4]
    exp = sp[0].header['EXPTIME']
    print(frame, exp)
```

```python
## first one is for frame 97-99, second one is for frame 102, 108 and 109
xtrim_list_det2 = [[855, 930],  [1070, 1145]] #[850, 935],  [1065, 1150]

xtrim_list_det6 = [[845, 930], [1065, 1140]]  #[840, 935], [1060, 1145]

spec_det_2_1 = []
spec_det_6_1 = []
spec_det_2_2 = []
spec_det_6_2 = []

for fl in file_list:
    frame = fl.split('_')[-3][:4]
    if int(frame) <100:
        #print(frame, xtrim_list_det2[0], xtrim_list_det6[0])
        spec1 = spec2d.Spec2d(fl, hext=14, xtrim=xtrim_list_det2[0])
        spec2 = spec2d.Spec2d(fl, hext=66, xtrim=xtrim_list_det6[0])
        
        spec1.set_dispaxis('y')
        spec1.subtract_sky_2d(use_skymod=True)
        spec_det_2_1.append(spec1.data)
    
        spec2.set_dispaxis('y')
        spec2.subtract_sky_2d(use_skymod=True)
        spec_det_6_1.append(spec2.data)
    
    else:
        #print(frame, xtrim_list_det2[1], xtrim_list_det6[1])
        spec1 = spec2d.Spec2d(fl, hext=14, xtrim=xtrim_list_det2[1])
        spec2 = spec2d.Spec2d(fl, hext=66, xtrim=xtrim_list_det6[1])
    
        spec1.set_dispaxis('y')
        spec1.subtract_sky_2d(use_skymod=True)
        spec_det_2_2.append(spec1.data)

        spec2.set_dispaxis('y')
        spec2.subtract_sky_2d(use_skymod=True)
        spec_det_6_2.append(spec2.data)
```

```python
median_im_det2_1 = np.median(np.asarray(spec_det_2_1), axis=0)
median_im_det2_2 = np.median(np.asarray(spec_det_2_2), axis=0)
median_im_det6_1 = np.median(np.asarray(spec_det_6_1), axis=0)
median_im_det6_2 = np.median(np.asarray(spec_det_6_2), axis=0)
```

```python
pf.PrimaryHDU(median_im_det2_1).writeto('J1817_DEIMOS_spectra/J1817_median_image_det2_1.fits')
pf.PrimaryHDU(median_im_det6_1).writeto('J1817_DEIMOS_spectra/J1817_median_image_det6_1.fits')
pf.PrimaryHDU(median_im_det2_2).writeto('J1817_DEIMOS_spectra/J1817_median_image_det2_2.fits')
pf.PrimaryHDU(median_im_det6_2).writeto('J1817_DEIMOS_spectra/J1817_median_image_det6_2.fits')
```

## subtract sky and cosmic rays

```python
extension = [14, 66]

xtrim_list = [[[855, 930], [845, 930]], [[1070, 1145], [1065, 1140]]]
#xtrim_list_det6 = [[840, 935], [1060, 1145]]

median_im = [['J1817_DEIMOS_spectra/J1817_median_image_det2_1.fits', 
             'J1817_DEIMOS_spectra/J1817_median_image_det6_1.fits',],
             ['J1817_DEIMOS_spectra/J1817_median_image_det2_2.fits',
             'J1817_DEIMOS_spectra/J1817_median_image_det6_2.fits']]

for fl, ofl1, ofl2 in zip(file_list, out_file_det2, out_file_det6):
    frame = fl.split('_')[-3][:4]
    if int(frame) <100:
        trim_list = xtrim_list[0]
        med_im = median_im[0]
    else:
        trim_list = xtrim_list[1]
        med_im = median_im[1]
        
    for ext, xtr, mdim in zip(extension, trim_list, med_im):
        print(ext, xtr, mdim)
        spec = spec2d.Spec2d(fl, hext=ext, xtrim=xtr)
        #spec2 = spec2d.Spec2d(fl, hext=66, xtrim=[970, 1080])
        spec.set_dispaxis('y')
        spec.subtract_sky_2d(use_skymod=True)
        spec.szap_new(infile=mdim)
    
        filled_data = spec.data
        mask = spec.szap_mask
        #print(filled_data.shape)
        ivar_im = pf.open(fl)[ext+1].data[spec.ymin:spec.ymax, spec.xmin:spec.xmax]
        sky_im = pf.open(fl)[ext+2].data[spec.ymin:spec.ymax, spec.xmin:spec.xmax]
        wav_im = pf.open(fl)[ext+7].data[spec.ymin:spec.ymax, spec.xmin:spec.xmax]
    
        ivar_im[mask] = 1.e-5
        #print(ivar_im.shape)
        
        prim_hdr = pf.open(fl)[0].header
        prim_hdr.append(('OBJECT', 'J1817+2729'))
        prim_hdr.append(('', 'Trimmed and cleaned data by Pritom'))
        phdu = pf.PrimaryHDU(header=prim_hdr)
        
        sci_hdr = pf.Header()
        sci_hdr.append(('OBJECT', 'J1817+2729'))
        sci_hdr.append(('IMTYPE', 'SCIENCE_IMAGE'))
        sci_im_hdu = pf.ImageHDU(data=filled_data, header=sci_hdr, name='SCI_IM')

        ivar_hdr = pf.Header()
        ivar_hdr.append(('IMTYPE', 'IVAR_IMAGE'))
        ivar_im_hdu = pf.ImageHDU(data=ivar_im, header=ivar_hdr, name='IVAR')

        sky_hdr = pf.Header()
        sky_hdr.append(('IMTYPE', 'SKY_MODEL'))
        sky_im_hdu = pf.ImageHDU(data=sky_im, header=sky_hdr, name='SKY_MOD')

        wav_hdr = pf.Header()
        wav_hdr.append(('IMTYPE', 'WAV_IMAGE'))
        wav_im_hdu = pf.ImageHDU(data=wav_im, header=wav_hdr, name='WAV_IM')

        mask_hdr = pf.Header()
        mask_hdr.append(('IMTYPE', 'MASK'))
        cmt = 'pixels masked as cosmic rays and filled with pixel values at same pixel'\
            'location from median image'
        mask_hdr.append(('', cmt))
        mask_hdu = pf.ImageHDU(data=mask, header=mask_hdr, name='MASK')

        hdul = pf.HDUList([phdu, sci_im_hdu, ivar_im_hdu, sky_im_hdu, wav_im_hdu, mask_hdu ])
        if ext==14:
            hdul.writeto(ofl1)
        else:
            hdul.writeto(ofl2)
```
