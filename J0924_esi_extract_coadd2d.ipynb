```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from keckcode_test.keckcode.esiredux import esi2d, esi1d
from specim_test.specim.specfuncs import spec2d, echelle1d, spec1d, echelle2d
```

```python
file_loc = 'J0924_ESI_spectra/*'
```

```python
spec2d_file_list = []
var2d_file_list = []
#out_file = []
for i, p in enumerate(glob.glob(file_loc)):
    file_name = p.split('/')[-1]
    if 'sp_order' in file_name:
        spec2d_file_list.append(p)
    elif 'var' in file_name:
        var2d_file_list.append(p)
```

```python
print(spec2d_file_list, var2d_file_list)
```

```python
spec_ord_4 = spec2d.Spec2d(inspec=spec2d_file_list[1], invar=var2d_file_list[3], logwav=True)
spec_ord_5 = spec2d.Spec2d(inspec=spec2d_file_list[3], invar=var2d_file_list[2], logwav=True)
spec_ord_6 = spec2d.Spec2d(inspec=spec2d_file_list[0], invar=var2d_file_list[0], logwav=True)
spec_ord_7 = spec2d.Spec2d(inspec=spec2d_file_list[2], invar=var2d_file_list[4], logwav=True)
spec_ord_8 = spec2d.Spec2d(inspec=spec2d_file_list[5], invar=var2d_file_list[1], logwav=True)
spec_ord_9 = spec2d.Spec2d(inspec=spec2d_file_list[4], invar=var2d_file_list[5], logwav=True)
```

```python
pixscale = np.array([0.137, 0.144, 0.149, 0.153, 0.158, 0.163])
dist = [0.9, 0.89]
for p in pixscale:
    print(dist/p)
```

```python
spec_list= [spec_ord_4, spec_ord_5, spec_ord_6, spec_ord_7, spec_ord_8, spec_ord_9]
order_list = [4, 5, 6, 7, 8, 9]

pixrange_list = [[2400, 2800], [2300, 2700], [2300, 2700], [2500, 2900], [1600, 2000], [1500, 1900]]

for i, sp in enumerate(spec_list):
    print('order : %d' %order_list[i])
    sp.spatial_profile(pixrange=pixrange_list[i])
    plt.show()
    sp.initial_model()
```

```python
def tie_pos1_moffat_4(model):
    return model.x_0_1 + (0.9 / pixscale[0]) #(0.886 / pixscale)

def tie_pos2_moffat_4(model):
    return model.x_0_1 + (1.79 / pixscale[0])  #(1.87 / pixscale) 

def tie_pos1_moffat_5(model):
    return model.x_0_1 + (0.9 / pixscale[1]) #(0.886 / pixscale)

def tie_pos2_moffat_5(model):
    return model.x_0_1 + (1.79 / pixscale[1])  #(1.87 / pixscale) 

def tie_pos1_moffat_6(model):
    return model.x_0_1 + (0.9 / pixscale[2]) #(0.886 / pixscale)

def tie_pos2_moffat_6(model):
    return model.x_0_1 + (1.79 / pixscale[2])  #(1.87 / pixscale) 

def tie_pos1_moffat_7(model):
    return model.x_0_1 + (0.9 / pixscale[3]) #(0.886 / pixscale)

def tie_pos2_moffat_7(model):
    return model.x_0_1 + (1.79 / pixscale[3])  #(1.87 / pixscale) 

def tie_pos1_moffat_8(model):
    return model.x_0_1 + (0.9 / pixscale[4]) #(0.886 / pixscale)

def tie_pos2_moffat_8(model):
    return model.x_0_1 + (1.79 / pixscale[4])  #(1.87 / pixscale) 

def tie_pos1_moffat_9(model):
    return model.x_0_1 + (0.9 / pixscale[5]) #(0.886 / pixscale)

def tie_pos2_moffat_9(model):
    return model.x_0_1 + (1.79 / pixscale[5])

def tie_gamma_moffat(model):
    return model.gamma_1

def tie_alpha_moffat(model):
    return model.alpha_1

func_dict = {'4_1' : tie_pos1_moffat_4, '4_2' : tie_pos2_moffat_4,
             '5_1' : tie_pos1_moffat_5, '5_2' : tie_pos2_moffat_5,
             '6_1' : tie_pos1_moffat_6, '6_2' : tie_pos2_moffat_6,
             '7_1' : tie_pos1_moffat_7, '7_2' : tie_pos2_moffat_7,
             '8_1' : tie_pos1_moffat_8, '8_2' : tie_pos2_moffat_8,
             '9_1' : tie_pos1_moffat_9, '9_2' : tie_pos2_moffat_9,
            }

pixscale = np.array([0.137, 0.144, 0.149, 0.153, 0.158, 0.163])
order_list = [4, 5, 6, 7, 8, 9]

for sp, od in zip(spec_list, order_list):
    if sp.mod0 is not None:
        print('order : %d' %od)
        sp.mod0[2].x_0.tied = func_dict['%d_%d' %(od, 1)] #tie_pos1_moffat
        sp.mod0[3].x_0.tied = func_dict['%d_%d' %(od, 2)] #tie_pos2_moffat
        sp.mod0[3].gamma.tied = tie_gamma_moffat
        sp.mod0[3].alpha.tied = tie_alpha_moffat
        sp.refined_model(sp.mod0)
```

```python
fitrange_list = np.array([[1100, 4100], [1000, 4100],  [800, 4100], [500, 4100], [0, 4100], [0, 4100]])

for i, sp in enumerate(spec_list):
    sp.find_and_trace(sp.mod0, stepsize=25, fitorder={'x_0_1':4, 'x_0_2':4, 'x_0_3':4},
                    fitrange=fitrange_list[i], polyorder=3)
    plt.show()
```

```python
extrange_list = np.array([[1200, -1], [1100, -1], [900, -1], [600, -1], [200, -1], [0, -1]])
                          
for sp, ext in zip(spec_list, extrange_list):
    sp.extract(method='modelfit', extrange=ext)
    plt.show()
```
