```python
def tie_pos1_moffat_5(model):
    return model.x_0_2 + (0.5 / spec.ordinfo[4]['pixscale']) #(0.686 / pixscale)

def tie_pos2_moffat_5(model):
    return model.x_0_2 + (0.6 / spec.ordinfo[4]['pixscale'])  #(1.87 / pixscale) 

def tie_pos3_moffat_5(model):
    return model.x_0_2 + (1.32 / spec.ordinfo[4]['pixscale'])

def tie_pos1_moffat_6(model):
    return model.x_0_2 + (0.5 / spec.ordinfo[5]['pixscale']) #(0.686 / pixscale)

def tie_pos2_moffat_6(model):
    return model.x_0_2 + (0.6 / spec.ordinfo[5]['pixscale'])  #(1.87 / pixscale) 

def tie_pos3_moffat_6(model):
    return model.x_0_2 + (1.32 / spec.ordinfo[5]['pixscale'])

def tie_pos1_moffat_7(model):
    return model.x_0_2 + (0.5 / spec.ordinfo[6]['pixscale']) #(0.686 / pixscale)

def tie_pos2_moffat_7(model):
    return model.x_0_2 + (0.6 / spec.ordinfo[6]['pixscale'])  #(1.87 / pixscale) 

def tie_pos3_moffat_7(model):
    return model.x_0_2 + (1.32 / spec.ordinfo[6]['pixscale'])

def tie_pos1_moffat_8(model):
    return model.x_0_2 + (0.5 / spec.ordinfo[7]['pixscale']) #(0.686 / pixscale)

def tie_pos2_moffat_8(model):
    return model.x_0_2 + (0.6 / spec.ordinfo[7]['pixscale'])  #(1.87 / pixscale) 

def tie_pos3_moffat_8(model):
    return model.x_0_2 + (1.32 / spec.ordinfo[7]['pixscale'])

def tie_pos1_moffat_9(model):
    return model.x_0_2 + (0.5 / spec.ordinfo[8]['pixscale']) #(0.686 / pixscale)

def tie_pos2_moffat_9(model):
    return model.x_0_2 + (0.6 / spec.ordinfo[8]['pixscale'])  #(1.87 / pixscale) 

def tie_pos3_moffat_9(model):
    return model.x_0_2 + (1.32 / spec.ordinfo[8]['pixscale'])

def tie_pos1_moffat_10(model):
    return model.x_0_2 + (0.5 / spec.ordinfo[9]['pixscale']) #(0.686 / pixscale)

def tie_pos2_moffat_10(model):
    return model.x_0_2 + (0.6 / spec.ordinfo[9]['pixscale'])  #(1.87 / pixscale) 

def tie_pos3_moffat_10(model):
    return model.x_0_2 + (1.32 / spec.ordinfo[9]['pixscale']) 

def tie_gamma_moffat(model):
    return model.gamma_4

def tie_alpha_moffat(model):
    return model.alpha_4

func_dict = {'5_1' : tie_pos1_moffat_5, '5_2' : tie_pos2_moffat_5, '5_3' : tie_pos3_moffat_5,
             '6_1' : tie_pos1_moffat_6, '6_2' : tie_pos2_moffat_6, '6_3' : tie_pos3_moffat_6,
             '7_1' : tie_pos1_moffat_7, '7_2' : tie_pos2_moffat_7, '7_3' : tie_pos3_moffat_7,
             '8_1' : tie_pos1_moffat_8, '8_2' : tie_pos2_moffat_8, '8_3' : tie_pos3_moffat_8,
             '9_1' : tie_pos1_moffat_9, '9_2' : tie_pos2_moffat_9, '9_3' : tie_pos3_moffat_9,
             '10_1' : tie_pos1_moffat_10, '10_2' : tie_pos2_moffat_10, '10_3' : tie_pos3_moffat_10 
            }

for sp, info in zip(spec[4:], spec.ordinfo[4:]):
    if sp.mod0 is not None:
        print(info['pixscale'])
        sp.mod0[1].mean.tied = func_dict['%d_%d' %(info['order'], 1)] #tie_pos1_moffat
        sp.mod0[3].x_0.tied = func_dict['%d_%d' %(info['order'], 2)] #tie_pos2_moffat
        sp.mod0[4].x_0.tied = func_dict['%d_%d' %(info['order'], 3)] #tie_pos2_moffat
        sp.mod0[3].gamma.tied = tie_gamma_moffat
        sp.mod0[3].alpha.tied = tie_alpha_moffat
        sp.mod0[2].gamma.tied = tie_gamma_moffat
        sp.mod0[2].alpha.tied = tie_alpha_moffat
        sp.refined_model(sp.mod0)
```
