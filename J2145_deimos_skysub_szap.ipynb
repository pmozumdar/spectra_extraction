```python
xtrim_det = {'0613_0100':[[1282, 1402], [1272, 1402]], '0613_0101':[[1282, 1402], [1272, 1402]],
             '0613_0105':[[1542, 1682], [1532, 1682]], '0613_0106':[[1542, 1682], [1532, 1682]],
             '0614_0105':[[1282, 1402], [1272, 1402]], '0615_0103':[[1210, 1350], [1200, 1350]]
            }
ext = {'0613_0105': [13, 61], '0613_0101':[14, 66], '0613_0106': [13, 61], '0613_0100': [14, 66], 
       '0614_0105': [14, 66], '0615_0103': [14, 66]}

spec_det_2_1 = []
spec_det_6_1 = []

spec_det_2_2 = []
spec_det_6_2 = []

for i, fl in enumerate(file_list):
    for dt in xtrim_det:
        if dt in fl:
            xtrim_dt = xtrim_det[dt]
            break
    spec1 = spec2d.Spec2d(fl, hext=ext[dt][0], varext=ext[dt][0]+1, xtrim=xtrim_dt[0], do_invar=True)
    spec2 = spec2d.Spec2d(fl, hext=ext[dt][1], varext=ext[dt][1]+1, xtrim=xtrim_dt[1], do_invar=True)
    
    spec1.set_dispaxis('y')
    spec1.subtract_sky_2d(use_skymod=True)
    
    spec2.set_dispaxis('y')
    spec2.subtract_sky_2d(use_skymod=True)
    
    ln = xtrim_dt[0][1] - xtrim_dt[0][0]
    if ln==120:
        spec_det_2_1.append(spec1.data)
        spec_det_6_1.append(spec2.data)
    else:
        spec_det_2_2.append(spec1.data)
        spec_det_6_2.append(spec2.data)
```

```python
wtsp = np.zeros(shifted_sp[0].shape)
wtsum = np.zeros(shifted_sp[0].shape)

for s, v in zip(shifted_sp, shifted_var):
    
    wt = np.zeros(wtsp.shape)
    mask = np.isnan(v) | (v<=0)
    wt[~mask] = 1.0 / v[~mask]
    #print(sum(sum(np.isnan(wt) | (wt<=0))))
    wtsp += wt * s
    #print(sum(sum(wtsp==0)))
    wtsum += wt
    #print(sum(sum(wtsum==0)))

wtsp[wtsum == 0] = 0
wtsum[wtsum == 0] = 1
outsp = wtsp / wtsum
outvar = 1.0 / wtsum 
```
