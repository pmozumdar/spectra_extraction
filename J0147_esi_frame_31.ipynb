```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from astropy.table import Table
from astropy.modeling import models
from keckcode_test.keckcode.esiredux import esi2d, esi1d
from specim_test.specim.specfuncs import spec2d, echelle1d, spec1d, echelle2d
#from specim.specfuncs import spec1d, specset1d
```

```python
file_loc = '../galaxy_spectra/ESI_data/*'
```

```python
spec2d_file_list = []
var2d_file_list = []
#out_file = []
for i,p in enumerate(glob.glob(file_loc)):
    if 'J0147' in p:
        if 'bgsub.fits' in p:
            spec2d_file_list.append(p)
        elif 'var.fits' in p:
            var2d_file_list.append(p)
            #frame = p.split('_')[-3][:4]
            #out_file.append('0053m1_r'+ frame + '_cleaned.fits')
        #print(p)
```

```python
spec = esi2d.Esi2d(spec2d_file_list[0], varfile=var2d_file_list[1])
print(spec2d_file_list[0])
print(var2d_file_list[1])
```

```python
pixrange = np.array([[2600, 3000], [2600, 3000], [2600, 3000], [1800, 2200], [1500, 1800]]) #(1900, 2400)
i = 0
for sp, info in zip(spec[5:], spec.ordinfo[5:]):
    print(info['pixscale'])
    title = 'Spatial profile for ' + str(info['name'])
    sp.spatial_profile(pixrange=pixrange[i], title=title)
    plt.show()
    sp.initial_model()
    i +=1
```

```python
def tie_pos1_moffat_6(model):
    return model.x_0_2 + (1.95 / spec.ordinfo[5]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_6(model):
    return model.x_0_2 + (3.15 / spec.ordinfo[5]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_7(model):
    return model.x_0_2 + (1.95 / spec.ordinfo[6]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_7(model):
    return model.x_0_2 + (3.15 / spec.ordinfo[6]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_8(model):
    return model.x_0_2 + (1.95 / spec.ordinfo[7]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_8(model):
    return model.x_0_2 + (3.15 / spec.ordinfo[7]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_9(model):
    return model.x_0_2 + (1.95 / spec.ordinfo[8]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_9(model):
    return model.x_0_2 + (3.15 / spec.ordinfo[8]['pixscale'])  #(1.87 / pixscale) 

def tie_pos1_moffat_10(model):
    return model.x_0_2 + (1.95 / spec.ordinfo[9]['pixscale']) #(0.886 / pixscale)

def tie_pos2_moffat_10(model):
    return model.x_0_2 + (3.15 / spec.ordinfo[9]['pixscale'])  #(1.87 / pixscale) 

def tie_gamma_moffat(model):
    return model.gamma_3

def tie_alpha_moffat(model):
    return model.alpha_3

func_dict = {
             '6_1' : tie_pos1_moffat_6, '6_2' : tie_pos2_moffat_6,
             '7_1' : tie_pos1_moffat_7, '7_2' : tie_pos2_moffat_7,
             '8_1' : tie_pos1_moffat_8, '8_2' : tie_pos2_moffat_8,
             '9_1' : tie_pos1_moffat_9, '9_2' : tie_pos2_moffat_9,
             '10_1' : tie_pos1_moffat_10, '10_2' : tie_pos2_moffat_10
            }

for sp, info in zip(spec, spec.ordinfo):
    if sp.mod0 is not None:
        print(info['pixscale'])
        sp.mod0[1].mean.tied = func_dict['%d_%d' %(info['order'], 1)] #tie_pos1_moffat
        sp.mod0[3].x_0.tied = func_dict['%d_%d' %(info['order'], 2)] #tie_pos2_moffat
        sp.mod0[2].gamma.tied = tie_gamma_moffat
        sp.mod0[2].alpha.tied = tie_alpha_moffat
        sp.refined_model(sp.mod0)
```

```python
fig = plt.figure(figsize=(25, 20))
spec.plot_fitted_model(columns=3, fig=fig)
```

```python
for sp in spec[5:]:
    sp.find_and_trace(sp.mod0, stepsize=25, fitorder={'stddev_1':3, 'mean_1':4, 'x_0_2':4, 'x_0_3':4},
                    fitrange=[0, 4100], polyorder=3)
    plt.show()
```

```python
extrange_list = np.array([[900, -1], [600, -1], [200, -1], [0, -1], [0, 2000]])

for sp, ext in zip(spec[5:], extrange_list):
    sp.extract(method='modelfit', extrange=ext)
    plt.show()
```

```python
outspec = []
orderinfo = np.array([6, 7, 8, 9, 10], dtype=[('order', int)])
for i in range(3):
    sp_l = []
    for sp in spec[5:]:
        sp_l.append(sp.spectra[i][1])
        #print(sp.spectra[i][0],'\n', sp.spectra[i][1])
    outspec.append(echelle1d.Ech1d(sp_l, ordinfo=orderinfo))
```
