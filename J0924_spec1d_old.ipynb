```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from specim_test.specim.specfuncs import spec2d, spec1d
```

```python
file_loc = './*'
```

```python
spec1d_primary_list_gauss = []
spec1d_secondary_list_gauss = []
spec1d_tertiary_list_gauss = []
spec1d_primary_list_moffat = []
spec1d_secondary_list_moffat = []
spec1d_tertiary_list_moffat = []
#out_file = []
for i,p in enumerate(glob.glob(file_loc)):
    if 'J0924' in p and 'spec1d' in p :
        if 'gauss' in p:
            if 'primary.fits' in p:
                spec1d_primary_list_gauss.append(p)
            elif 'secondary.fits' in p:
                spec1d_secondary_list_gauss.append(p)
            elif 'tertiary.fits' in p:
                spec1d_tertiary_list_gauss.append(p)
        else:
            if 'primary.fits' in p:
                spec1d_primary_list_moffat.append(p)
            elif 'secondary.fits' in p:
                spec1d_secondary_list_moffat.append(p)
            elif 'tertiary.fits' in p:
                spec1d_tertiary_list_moffat.append(p)
```

```python
print(spec1d_primary_list_gauss)
print(spec1d_secondary_list_gauss)
print(spec1d_tertiary_list_gauss)
print(spec1d_primary_list_moffat)
print(spec1d_secondary_list_moffat)
print(spec1d_tertiary_list_moffat)
```
