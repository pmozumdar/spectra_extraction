```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from keckcode_test.keckcode.esiredux import esi2d, esi1d
from specim_test.specim.specfuncs import spec2d, spec1d, echelle1d, echelle2d
```

```python
file_loc = 'J0248_ESI_spectra/*'
```

```python
spec2d_file_list = []
var2d_file_list = []
#out_file = []
for i, p in enumerate(glob.glob(file_loc)):
    file_name = p.split('/')[-1]
    if 'sp' in file_name and 'order' in file_name:
        spec2d_file_list.append(p)
    elif 'var' in file_name:
        var2d_file_list.append(p)
```

```python
print(spec2d_file_list, var2d_file_list)
```

```python
spec_ord_5 = spec2d.Spec2d(inspec=spec2d_file_list[4], invar=var2d_file_list[2], logwav=True)
spec_ord_6 = spec2d.Spec2d(inspec=spec2d_file_list[1], invar=var2d_file_list[3], logwav=True)
spec_ord_7 = spec2d.Spec2d(inspec=spec2d_file_list[0], invar=var2d_file_list[0], logwav=True)
spec_ord_8 = spec2d.Spec2d(inspec=spec2d_file_list[5], invar=var2d_file_list[5], logwav=True)
spec_ord_9 = spec2d.Spec2d(inspec=spec2d_file_list[2], invar=var2d_file_list[1], logwav=True)
spec_ord_10 = spec2d.Spec2d(inspec=spec2d_file_list[3], invar=var2d_file_list[4], logwav=True)
```

```python
spec_list= [spec_ord_5, spec_ord_6, spec_ord_7, spec_ord_8, spec_ord_9, spec_ord_10]
pixrange_list = [[2200, 2600], [2200, 2600], [1800, 2200], [1500, 1900], [1500, 1900], [1500, 1900]]
pixscale = [0.144, 0.149, 0.153, 0.158, 0.163, 0.168]
for i, sp in enumerate(spec_list):
    print(pixscale[i])
    sp.spatial_profile(pixrange=pixrange_list[i])
    plt.show()
    sp.initial_model()
```

```python
def tie_pos1_moffat_5(model):
    return model.x_0_3 - (0.63 / pixscale[0]) #(0.886 / pixscale)

def tie_pos2_moffat_5(model):
    return model.x_0_3 - (1.04 / pixscale[0])  #(1.87 / pixscale) 

def tie_pos1_moffat_6(model):
    return model.x_0_3 - (0.63 / pixscale[1]) #(0.886 / pixscale)

def tie_pos2_moffat_6(model):
    return model.x_0_3 - (1.04 / pixscale[1])

def tie_pos1_moffat_7(model):
    return model.x_0_3 - (0.63 / pixscale[2]) #(0.886 / pixscale)

def tie_pos2_moffat_7(model):
    return model.x_0_3 - (1.04 / pixscale[2])  #(1.87 / pixscale) 

def tie_pos1_moffat_8(model):
    return model.x_0_3 - (0.63 /  pixscale[3]) #(0.886 / pixscale)

def tie_pos2_moffat_8(model):
    return model.x_0_3 - (1.04 /  pixscale[3])  #(1.87 / pixscale) 

def tie_pos1_moffat_9(model):
    return model.x_0_3 - (0.63 /  pixscale[4]) #(0.886 / pixscale)

def tie_pos2_moffat_9(model):
    return model.x_0_3 - (1.04 /  pixscale[4])  #(1.87 / pixscale) 

def tie_pos1_moffat_10(model):
    return model.x_0_3 - (0.63 /  pixscale[5]) #(0.886 / pixscale)

def tie_pos2_moffat_10(model):
    return model.x_0_3 - (1.04 /  pixscale[5])  #(1.87 / pixscale) 

def tie_gamma_moffat1(model):
    return model.gamma_3

def tie_alpha_moffat1(model):
    return model.alpha_3

def tie_gamma_moffat2(model):
    return model.gamma_3 

def tie_alpha_moffat2(model):
    return model.alpha_3

func_dict = {'5_1' : tie_pos1_moffat_5, '5_2' : tie_pos2_moffat_5,
             '6_1' : tie_pos1_moffat_6, '6_2' : tie_pos2_moffat_6,
             '7_1' : tie_pos1_moffat_7, '7_2' : tie_pos2_moffat_7,
             '8_1' : tie_pos1_moffat_8, '8_2' : tie_pos2_moffat_8,
             '9_1' : tie_pos1_moffat_9, '9_2' : tie_pos2_moffat_9,
             '10_1' : tie_pos1_moffat_10, '10_2' : tie_pos2_moffat_10
            }
pixscale = [0.144, 0.149, 0.153, 0.158, 0.163, 0.168]
ord_list = [5, 6, 7, 8, 9, 10]
for i, sp in enumerate(spec_list):
    if sp.mod0 is not None:
        print(pixscale[i])
        sp.mod0[2].x_0.tied = func_dict['%d_%d' %(ord_list[i], 1)] #tie_pos1_moffat
        sp.mod0[1].x_0.tied = func_dict['%d_%d' %(ord_list[i], 2)] #tie_pos2_moffat
        sp.mod0[1].gamma.tied = tie_gamma_moffat1
        sp.mod0[1].alpha.tied = tie_alpha_moffat1
        #sp.mod0[2].gamma.tied = tie_gamma_moffat2
        #sp.mod0[2].alpha.tied = tie_alpha_moffat2
        sp.refined_model(sp.mod0)
```

```python
fitrange_list = np.array([[1000, 4100], [1000, 4100], [0, 4100],
                          [0, 4100], [0, 4100], [0, 4100]])
for i, sp in enumerate(spec_list):
    sp.find_and_trace(sp.mod0, stepsize=25, fitorder={'x_0_1':4, 'mean_1':4, 'x_0_2':4, 'x_0_3':4},
                    fitrange=fitrange_list[i], polyorder=3)
    plt.show()
```

```python
extrange_list = np.array([[1100, -1], [900, -1], [600, -1], 
                          [200, -1], [0, -1], [0, 2000]])
for sp, ext in zip(spec_list, extrange_list):
    sp.extract(method='modelfit', extrange=ext)
    plt.show()
```
