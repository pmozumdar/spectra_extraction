```python
spec = esi2d.Esi2d(spec2d_loc, var2d_loc)
```

```python
pixrange_list = [[1700, 2100], [1600, 2000], [800, 1200]] 

for i, sp in enumerate(spec[7:]):
    sp.spatial_profile(pixrange=pixrange_list[i])
    plt.show()
    sp.initial_model()
```

```python
def tie_pos1_moffat_8(model):
    return model.x_0_1 + (1.75 /  pixscale[0])  

def tie_pos2_moffat_8(model):
    return model.x_0_1 + (0.80 /  pixscale[0])

def tie_pos3_moffat_8(model):
    return model.x_0_1 + (1.65 / pixscale[0])

def tie_pos1_moffat_9(model):
    return model.x_0_1 + (1.75 /  pixscale[1])  

def tie_pos2_moffat_9(model):
    return model.x_0_1 + (0.80 /  pixscale[1])

def tie_pos3_moffat_9(model):
    return model.x_0_1 + (1.65 / pixscale[1])

def tie_pos1_moffat_10(model):
    return model.x_0_1 + (1.75 /  pixscale[2])  

def tie_pos2_moffat_10(model):
    return model.x_0_1 + (0.80 /  pixscale[2])

def tie_pos3_moffat_10(model):
    return model.x_0_1 + (1.65 / pixscale[2])

def tie_gamma_moffat(model):
    return model.gamma_1

def tie_alpha_moffat(model):
    return model.alpha_1 

def tie_gamma_moffat1(model):
    return model.gamma_1 * 0.5

def tie_alpha_moffat1(model):
    return model.alpha_1 * 0.5

func_dict = {
 '8_1' : tie_pos1_moffat_8, '8_2' : tie_pos2_moffat_8, '8_3' : tie_pos3_moffat_8,
 '9_1' : tie_pos1_moffat_9, '9_2' : tie_pos2_moffat_9, '9_3' : tie_pos3_moffat_9,
 '10_1' : tie_pos1_moffat_10, '10_2' : tie_pos2_moffat_10, '10_3' : tie_pos3_moffat_10
}
pixscale = [0.158, 0.163, 0.168]
ord_list = [8, 9, 10]
for i, sp in enumerate(spec[7:]):
    if sp.mod0 is not None:
        print(pixscale[i])
        #sp.mod0[3].x_0.tied = func_dict['%d_%d' %(ord_list[i], 2)] #tie_pos1_moffat
        #sp.mod0[1].mean.tied = func_dict['%d_%d' %(ord_list[i], 3)] #tie_pos1_moffat
        sp.mod0[2].x_0.tied = func_dict['%d_%d' %(ord_list[i], 2)] #tie_pos2_moffat
        sp.mod0[3].x_0.tied = func_dict['%d_%d' %(ord_list[i], 1)]
        #sp.mod0[5].x_0.tied = func_dict['%d_%d' %(ord_list[i], 3)]
        sp.mod0[2].gamma.tied = tie_gamma_moffat
        sp.mod0[2].alpha.tied = tie_alpha_moffat
        sp.mod0[3].gamma.tied = tie_gamma_moffat
        sp.mod0[3].alpha.tied = tie_alpha_moffat
        #sp.mod0[4].gamma.tied = tie_gamma_moffat
        #sp.mod0[4].alpha.tied = tie_alpha_moffat
        #sp.mod0[5].gamma.tied = tie_gamma_moffat
        #sp.mod0[5].alpha.tied = tie_alpha_moffat
        sp.refined_model(sp.mod0)
```

```python
fitrange_list = np.array([[0, 4100], [0, 4100], [0, 4100]])
for i, sp in enumerate(spec[7:]):
    sp.find_and_trace(sp.mod0, stepsize=25, fitorder={'x_0_1':4, 'mean_1':4, 'x_0_2':4, 'x_0_3':4},
                    fitrange=fitrange_list[i], polyorder=3)
    plt.show()
```

```python
extrange_list = np.array([[200, -1], [0, -1], [0, 2500]])
for sp, ext in zip(spec[7:], extrange_list):
    sp.extract(method='modelfit', extrange=ext)
    plt.show()
```
