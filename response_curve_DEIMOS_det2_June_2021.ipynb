### Create response curves

```python
obs_spec1 = spec1d.Spec1d('standard_star_DEIMOS_2021/spec1d_Feige66_DEIMOS_det3_June13_2021.fits',
                          informat='fits')
obs_spec2 = spec1d.Spec1d('standard_star_DEIMOS_2021/spec1d_Feige66_DEIMOS_det3_June14_2021.fits',
                          informat='fits')
```

```python
%matplotlib notebook
obs_spec1.plot()
```

### Each element in the list contains crude wavelength range of the region to mask and the width in the blue and red side of the region to use for continum fitting.

```python
linelist1 = [(4675, 4730, 40), (4810, 4950, 200, 60), (4995, 5060, 50), (5845, 5910, 100),
            (6260, 6305, 100), (6520, 6620, 150, 50), (6665, 6695, 50), (6855, 6950, 100), (7055, 7075, 25)]
#(5145, 5290, 150),
```

```python
clst_wav_range, wav_index = obs_spec1.closest_wavrange(linelist1, verbose=False)
for i, l in enumerate(linelist1): 
    try:
        r = [l[2], l[3]]
    except:
        r = [l[2], l[2]]
    obs_spec1.mask_line(clst_wav_range[i], r)
```

```python
obs_spec1.plot()
```

```python
obs_spec1.smooth(5)
```

```python
calb_spec = spec1d.Spec1d('../other_spectra/feige110_ir_osp.fits', informat='fitsflux')
```

```python
modlines = [(3700.46, 5020, 100, 600), (6524.47, 6619.27, 10, 10)]
clst_wav_range1, wav_index1 = calb_spec.closest_wavrange(modlines, verbose=False)

for l, w in zip(modlines, clst_wav_range1):
    calb_spec.mask_line(w, [l[2], l[3]])
calb_spec.plot()
```

```python
calb_spec.resample(owave=obs_spec1['wav'])
```

```python
response1 = obs_spec1['flux'] / (calb_spec.rsflux / ((np.median(calb_spec.rsflux)/10**3)))
```

```python
plt.figure()
plt.plot(obs_spec1['wav'], response1)
plt.title('response curve')
```

```python
response1_1d = spec1d.Spec1d(wav=obs_spec1['wav'], flux=response1)
```

```python
response1_1d.save('standard_star_DEIMOS_2021/response_curve_DEIMOS_det3_June13_2021.txt', outformat='text')
```
