```python
import numpy as np
import matplotlib.pyplot as plt
import glob

from specim_test.specim.specfuncs import spec2d, spec1d
```

```python
file_loc = 'J0147_spectra/*'
```

```python
spec1d_primary_list= []
spec1d_secondary_list = []
spec1d_deflector_list = []
#spec1d_primary_list_moffat = []
#spec1d_secondary_list_moffat = []
#spec1d_tertiary_list_moffat = []
#out_file = []
for i,p in enumerate(glob.glob(file_loc)):
    if 'J0147' in p and 'spec1d' in p :
        if 'frame' in p:
            if 'primary.fits' in p:
                spec1d_primary_list.append(p)
            elif 'secondary.fits' in p:
                spec1d_secondary_list.append(p)
            elif 'deflector.fits' in p:
                spec1d_deflector_list.append(p)
        #else:
            #if 'primary.fits' in p:
            #    spec1d_primary_list_moffat.append(p)
            #elif 'secondary.fits' in p:
            #    spec1d_secondary_list_moffat.append(p)
            #elif 'tertiary.fits' in p:
            #    spec1d_tertiary_list_moffat.append(p)
```

```python
for i, p in enumerate(spec1d_deflector_list):
    if i==0:
        sp_d = spec1d.Spec1d(p, informat='fits')
        sp_d.plot()
    else:
        sp1 = spec1d.Spec1d(p, informat='fits')
        sp1.plot()
        sp_d = sp_d.__add__(sp1)
        sp_d.plot()
```
