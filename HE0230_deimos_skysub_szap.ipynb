```python
xtrim_det = [[398, 591], [652, 845]]

spec_det_3 = []
spec_det_7 = []

for i, fl in enumerate(file_list):
    
    spec1 = spec2d.Spec2d(fl, hext=31, varext=32, do_invar=True)
    spec2 = spec2d.Spec2d(fl, hext=91, varext=92, do_invar=True)
    
    spec1.set_dispaxis('y')
    spec1.subtract_sky_2d(use_skymod=True)
    
    spec2.set_dispaxis('y')
    spec2.subtract_sky_2d(use_skymod=True)
    
    if i<3:
        spec_det_3.append(spec1.data[:, xtrim_det[0][0]:xtrim_det[0][1]])
        spec_det_7.append(spec2.data[:, xtrim_det[0][0]:xtrim_det[0][1]])
    else:
        spec_det_3.append(spec1.data[:, xtrim_det[1][0]:xtrim_det[1][1]])
        spec_det_7.append(spec2.data[:, xtrim_det[1][0]:xtrim_det[1][1]])   
```

```python
median_im_det3 = np.median(np.asarray(spec_det_3), axis=0)
median_im_det7 = np.median(np.asarray(spec_det_7), axis=0)
```

```python
folder_path = 'HE0230_DEIMOS_spectra' 
# Create the folder
try:
    os.makedirs(folder_path)
except:
    print(f'folder already exist for {folder_path}')
    pass
pf.PrimaryHDU(median_im_det3).writeto('HE0230_DEIMOS_spectra/HE0230_median_image_det3.fits', overwrite=True)
pf.PrimaryHDU(median_im_det7).writeto('HE0230_DEIMOS_spectra/HE0230_median_image_det7.fits', overwrite=True)
```

### subtract sky and cosmic rays

```python
extension = [31, 91]
trim = [[398, 591], [652, 845]]

median_im = ['HE0230_DEIMOS_spectra/HE0230_median_image_det3.fits', 
             'HE0230_DEIMOS_spectra/HE0230_median_image_det7.fits']
i=0
for fl, ofl1, ofl2 in zip(file_list, out_file_det3, out_file_det7):
    if i<3:
         x1, x2 = xtrim_det[0][0], xtrim_det[0][1]
    else:
        x1, x2 = xtrim_det[1][0], xtrim_det[1][1]
        
    for ext, mdim in zip(extension, median_im):
        print(ext, mdim)
        spec = spec2d.Spec2d(fl, hext=ext, varext=ext+1)
        #spec2 = spec2d.Spec2d(fl, hext=66, xtrim=[970, 1080])
        spec.set_dispaxis('y')
        spec.subtract_sky_2d(use_skymod=True)
        spec.data = spec.data[:, x1:x2]
        spec.szap_new(infile=mdim)

        if ext==31:
            filled_data = spec.data
            mask = spec.szap_mask
            #print(filled_data.shape)
            ivar_im = pf.open(fl)[ext+1].data[spec.ymin:spec.ymax, spec.xmin:spec.xmax][:, x1:x2]
            sky_im = pf.open(fl)[ext+2].data[spec.ymin:spec.ymax, spec.xmin:spec.xmax][:, x1:x2]
            wav_im = pf.open(fl)[ext+7].data[spec.ymin:spec.ymax, spec.xmin:spec.xmax][:, x1:x2]

        elif ext==91:
            filled_data = spec.data
            mask = spec.szap_mask
            #print(filled_data.shape)
            ivar_im = pf.open(fl)[ext+1].data[spec.ymin:spec.ymax, spec.xmin:spec.xmax][:, x1:x2]
            sky_im = pf.open(fl)[ext+2].data[spec.ymin:spec.ymax, spec.xmin:spec.xmax][:, x1:x2]
            wav_im = pf.open(fl)[ext+7].data[spec.ymin:spec.ymax, spec.xmin:spec.xmax][:, x1:x2]

        ivar_im[mask] = 1.e-5
        #print(ivar_im.shape)

        prim_hdr = pf.open(fl)[0].header
        prim_hdr.append(('OBJECT', 'HE0230-2130'))
        prim_hdr.append(('', 'Trimmed and cleaned data by Pritom'))
        phdu = pf.PrimaryHDU(header=prim_hdr)

        sci_hdr = pf.Header()
        sci_hdr.append(('OBJECT', 'HE0230-2130'))
        sci_hdr.append(('IMTYPE', 'SCIENCE_IMAGE'))
        sci_im_hdu = pf.ImageHDU(data=filled_data, header=sci_hdr, name='SCI_IM')

        ivar_hdr = pf.Header()
        ivar_hdr.append(('IMTYPE', 'IVAR_IMAGE'))
        ivar_im_hdu = pf.ImageHDU(data=ivar_im, header=ivar_hdr, name='IVAR')

        sky_hdr = pf.Header()
        sky_hdr.append(('IMTYPE', 'SKY_MODEL'))
        sky_im_hdu = pf.ImageHDU(data=sky_im, header=sky_hdr, name='SKY_MOD')

        wav_hdr = pf.Header()
        wav_hdr.append(('IMTYPE', 'WAV_IMAGE'))
        wav_im_hdu = pf.ImageHDU(data=wav_im, header=wav_hdr, name='WAV_IM')

        mask_hdr = pf.Header()
        mask_hdr.append(('IMTYPE', 'MASK'))
        cmt = 'pixels masked as cosmic rays and filled with pixel values at same pixel'\
            'location from median image'
        mask_hdr.append(('', cmt))
        mask_hdu = pf.ImageHDU(data=mask, header=mask_hdr, name='MASK')

        hdul = pf.HDUList([phdu, sci_im_hdu, ivar_im_hdu, sky_im_hdu, wav_im_hdu, mask_hdu ])

        if ext==31:
            hdul.writeto(ofl1)
            print(ofl1)
        else:
            hdul.writeto(ofl2)
            print(ofl2)
    i +=1
```
